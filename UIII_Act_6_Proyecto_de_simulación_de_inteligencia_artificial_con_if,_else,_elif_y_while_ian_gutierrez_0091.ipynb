##26. Cuenta de ahorros

```python
# ===================================
# CUENTA DE AHORROS
# ===================================

print("===================================")
print(" CUENTA DE AHORROS ")
print("===================================")

nombre = input("Ingrese el nombre del cliente: ")
saldo = float(input("Ingrese el saldo inicial: "))

print("\nANALIZANDO CUENTA...\n")

# -----------------------------------
# ANALISIS PRINCIPAL
# -----------------------------------

if saldo >= 50000:
    print("Excelente ahorro.")
    print("Cuenta con saldo muy alto.")

elif saldo >= 20000:
    print("Buen nivel de ahorro.")
    print("Cuenta estable.")

elif saldo >= 5000:
    print("Ahorro medio.")
    print("Puede seguir ahorrando.")

elif saldo > 0:
    print("Saldo bajo.")
    print("Se recomienda ahorrar mas.")

else:
    print("Cuenta sin fondos.")
    print("Necesita realizar un deposito.")

# -----------------------------------
# ANALISIS EXTRA
# -----------------------------------

prestamo = input("¿Tiene prestamos pendientes? (si/no): ")
tarjeta = input("¿Tiene tarjeta de credito? (si/no): ")

if prestamo == "si":
    print("Advertencia: controle sus deudas.")

if tarjeta == "si":
    print("Cliente con servicios bancarios activos.")

# -----------------------------------
# REPORTE FINAL
# -----------------------------------

print("\n===================================")
print(" REPORTE FINAL ")
print("===================================")

print("Cliente:", nombre)
print("Saldo:", saldo)
print("programa realizado por ian gutierrez NL=26")

print("\nGracias por utilizar CUENTA DE AHORROS")
```
