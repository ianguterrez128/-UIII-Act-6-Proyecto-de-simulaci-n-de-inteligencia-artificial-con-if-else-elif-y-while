##26. Cuenta de ahorros

```python
# =========================================
# CUENTA DE AHORROS
# Uso de if, elif, else y while
# =========================================

print("===================================")
print(" CUENTA DE AHORROS ")
print("===================================")

nombre = input("Ingrese el nombre del cliente: ")
saldo = float(input("Ingrese el saldo inicial: "))

opcion = 0

while opcion != 5:

    print("\n=========== MENU ===========")
    print("1. Consultar saldo")
    print("2. Depositar dinero")
    print("3. Retirar dinero")
    print("4. Analizar cuenta")
    print("5. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # ---------------------------------
    # CONSULTAR SALDO
    # ---------------------------------
    if opcion == 1:
        print("\nCliente:", nombre)
        print("Saldo actual:", saldo)

    # ---------------------------------
    # DEPOSITAR
    # ---------------------------------
    elif opcion == 2:

        deposito = float(input("Cantidad a depositar: "))

        if deposito > 0:
            saldo = saldo + deposito
            print("Deposito realizado correctamente")
            print("Nuevo saldo:", saldo)

        else:
            print("Cantidad invalida")

    # ---------------------------------
    # RETIRAR
    # ---------------------------------
    elif opcion == 3:

        retiro = float(input("Cantidad a retirar: "))

        if retiro <= saldo and retiro > 0:
            saldo = saldo - retiro
            print("Retiro realizado correctamente")
            print("Saldo restante:", saldo)

        elif retiro > saldo:
            print("Fondos insuficientes")

        else:
            print("Cantidad invalida")

    # ---------------------------------
    # ANALISIS
    # ---------------------------------
    elif opcion == 4:

        print("\nANALIZANDO CUENTA...")

        if saldo >= 10000:
            print("Excelente ahorro.")
            print("Cuenta con saldo alto.")

        elif saldo >= 5000 and saldo < 10000:
            print("Buen nivel de ahorro.")

        elif saldo >= 1000 and saldo < 5000:
            print("Ahorro medio.")

        elif saldo > 0 and saldo < 1000:
            print("Saldo bajo.")
            print("Se recomienda ahorrar mas.")

        else:
            print("Cuenta sin fondos.")

    # ---------------------------------
    # SALIR
    # ---------------------------------
    elif opcion == 5:
        print("\n===================================")
        print(" REPORTE FINAL ")
        print("===================================")
        print("Cliente:", nombre)
        print("Saldo final:", saldo)
        print("Programa realizado por Ian Gutierrez NL=26")
        print("Gracias por utilizar CUENTA DE AHORROS")

    # ---------------------------------
    # OPCION INVALIDA
    # ---------------------------------
    else:
        print("Opcion no valida")
```
